```python
import math

def sin(x):
    return math.sin(x)
def cos(x):
    return math.cos(x)
#had to rewrite these

start = 0.0
end = 2.0
num_points = 1000
step = (end - start) / (num_points - 1)
x_values = []
sin_values = []
cos_values = []

for i in range(num_points):
    x = start + i * step
    x_values.append(x)
    sin_values.append(sin(x))
    cos_values.append(cos(x))

print("   x\t\t sin(x)\t\t cos(x)")
print("---------------------------------------")

for i in range(10):
    print(f"{x_values[i]:.6f}\t {sin_values[i]:.6f}\t {cos_values[i]:.6f}")
```

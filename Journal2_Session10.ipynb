```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.exponential(scale=1.0, size=1000)

plt.figure()
plt.hist(random_numbers, bins=100)

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.title("Histogram of Exponential Distribution")

plt.savefig("exponential_histogram.pdf")

plt.show()
```

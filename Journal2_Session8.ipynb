```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)

def my_sin(x):
    return np.sin(x)

def my_cos(x):
    return np.cos(x)
```

```python

y_sin = my_sin(x)
y_cos = my_cos(x)

fig, axes = plt.subplots(1, 2)

axes[0].plot(x, y_sin)
axes[0].set_title("sin(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

axes[1].plot(x, y_cos)
axes[1].set_title("cos(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

plt.savefig("trig_multipanel.pdf")

plt.show()
```
